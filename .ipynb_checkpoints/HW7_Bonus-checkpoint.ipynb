```python
import pandas as pd
from sqlalchemy import create_engine
from dbconfig import url_path
import matplotlib.pyplot as plt

#Create engine and connection to DB through .py file using load_dotenv()
engine = create_engine(url_path)
conn = engine.connect()
```

```python
#Import tables as data frames

employees = pd.read_sql("SELECT * FROM employees", conn)
departments = pd.read_sql("SELECT * FROM departments", conn)
dept_emp = pd.read_sql("SELECT * FROM dept_emp", conn)
dept_manager = pd.read_sql("SELECT * FROM dept_manager", conn)
titles = pd.read_sql("SELECT * FROM titles", conn)
salaries = pd.read_sql("SELECT * FROM salaries", conn)
```

```python
#Create a histogram of salaries
hist = salaries.hist(column='salary')
```

```python
#find averages of different titles and create bar graph
ts_df = salaries.merge(titles, on='emp_no')

avg_salary_df = pd.DataFrame()
avg_salary_df['Avg Salary'] = ts_df.groupby('title')['salary'].mean().round(2)

avg_salary_df.plot.barh(title='Average Salary for each Title',legend=False)
plt.xlabel('Average Salary')
plt.ylabel('Title')
plt.show()
```
